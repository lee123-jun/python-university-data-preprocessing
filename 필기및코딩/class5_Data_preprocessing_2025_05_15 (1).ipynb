```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:\\Users\\User\\Desktop\\Data_Sample.csv")
data
```

```python
# 1. 결측치 개수 도출
data.isnull().sum()
```

```python
# 2. 칼럼 이름 변경
data = data.rename(columns={'Value1':'A'})
data = data.rename(columns={'Value2':'B'})
data = data.rename(columns={'Value3':'C'})
data
```

```python
# 3. 범주형 연속형 분리, 결측값 중앙값 대치
A=data.select_dtypes(include=["number"]).columns
A
```

```python
B=data.select_dtypes(include=["object", "category"]).columns
B
```

```python
data[A] = data[A].fillna(data[A].median())
data[A].isnull().sum()
```

```python
# 4. 하나로 다시 합치기
pd.concat([data[A], data[B]], axis=1) # 행기준
```

```python
data.isnull().sum()
```

```python
data
```

```python
# 5. A+B+C = Total 하나로 합치고 기존 a,b,c 삭제
total = data['A'] + data['B'] + data['C']
total
```

```python
data = data.drop(['A', 'B', 'C'], axis=1)
```

```python
data['total'] = total
data
```

```python
# 6. model_serial 두개의 변수명으로 나누기
split_data = data['Model_Serial'].str.split('-', expand=True)
split_data = split_data.rename(columns={0:'Model'})
split_data = split_data.rename(columns={1:'Serial'})
```

```python
# 7. 잘못된 값 수정
split_data['Model'].unique()
```

```python
split_data.replace('Gamma ','Gamma')
```

```python
split_data['Model']=split_data['Model'].str.replace(r'Gamma ','Gamma', regex=True)
split_data['Model']=split_data['Model'].str.replace(r' Beta','Beta', regex=True)
split_data['Model']=split_data['Model'].str.replace(r' Alpha','A', regex=True)
split_data['Model'].unique()
```
