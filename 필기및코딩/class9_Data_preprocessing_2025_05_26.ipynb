```python
H, W = 10,10
R = np.zeros((H,W), dtype=np.uint8)
R[0:5, 0:5] = 255

# 1. R  (Grayscale)
plt.figure(figsize=(20, 4))
plt.imshow(R, cmap='gray')
plt.title('R-Channel')
plt.show()
```

```python
#    
G = np.zeros((H, W), dtype=np.uint8)
G[3:7, 3:7] = 255
# 2. G  (Grayscale)
plt.figure(figsize=(20, 4))
plt.imshow(G, cmap='gray')
plt.title('G-Channel')
plt.show()
```

```python
#    
B = np.zeros((H, W), dtype=np.uint8)
B[5:10, 5:10] = 255
# 3. B  (Grayscale)
plt.figure(figsize=(20, 4))
plt.imshow(B, cmap='gray')
plt.title('B-Channel')
plt.show()
```

```python
#     
two_channel = np.stack([R, G, np.zeros_like(B)], axis=-1)
# 4. 2 (R+G   )
plt.figure(figsize=(20, 4))
plt.imshow(two_channel)
plt.title('R + G')
plt.show()
```

```python
#     
three_channel = np.stack([R, G, B], axis=-1)
# 5. 3 (RGB  )
plt.figure(figsize=(20, 4))
plt.imshow(three_channel)
plt.title('R + G + B')
plt.show()
```

```python
# R, G, B  
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(RGB_image)
```

```python
#  
plt.subplot(1,3,1)
plt.imshow(R, cmap='Reds')
plt.title('Red Channel')

#  
plt.subplot(1,3,2)
plt.imshow(G, cmap='Greens')
plt.title('Green Channel')

#  
plt.subplot(1,3,3)
plt.imshow(B, cmap='Blues')
plt.title('Blue Channel')

plt.show()
```

```python
#   BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#   RGB to HSV
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
```

```python
# H, S, V  
H, S, V = cv2.split(image_hsv)
```

```python
#     
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)

#  
plt.imshow(H, cmap= 'hsv')
plt.title('Hue Channe')

# 
plt.subplot(1, 3, 2)
plt.imshow(S, cmap= 'gray')
plt.title('Saturation Channel')

# 
plt.subplot(1, 3, 3)
plt.imshow(V, cmap= 'gray')
plt.title('Value Channel') 
plt.show()
```

```python
#  64x64  
# 10*10          !
resized_image=cv2.resize(image, (20,20), interpolation=cv2.INTER_AREA)
resized_image=cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
```

```python
#    5*5
blurred_image = cv2.GaussianBlur(image, (5,5),0)
blurred_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)
```

```python
# 
median_filtered_img = cv2.medianBlur(image, 5)
median_filtered_img = cv2.cvtColor(median_filtered_img, cv2.COLOR_BGR2RGB)
```

```python
#    
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.imshow(blurred_image)
plt.title('Blurred')

plt.subplot(1,2,2)
plt.imshow(median_filtered_img)
plt.title('Median')

plt.show()
```
