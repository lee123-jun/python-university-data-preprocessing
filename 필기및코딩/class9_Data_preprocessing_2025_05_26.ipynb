```python
# 무작위 생성으로 이미지 만들기 rgb
random_image = np.random.randint(0, 256, (100,100), dtype=np.uint8)
RGB_image = cv2.cvtColor(random_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,4))
plt.imshow(RGB_image)
plt.show()
```

```python
# 해당 이미지의 수치 값
RGB_image
```

```python
H, W = 10,10
R = np.zeros((H,W), dtype=np.uint8)
R[0:5, 0:5] = 255

# 1. R 채널 (Grayscale)
plt.figure(figsize=(20, 4))
plt.imshow(R, cmap='gray')
plt.title('R-Channel')
plt.show()
```

```python
# 배열을 활용해 이미지 생성
G = np.zeros((H, W), dtype=np.uint8)
G[3:7, 3:7] = 255
# 2. G 채널 (Grayscale)
plt.figure(figsize=(20, 4))
plt.imshow(G, cmap='gray')
plt.title('G-Channel')
plt.show()
```

```python
# 배열을 활용해 이미지 생성
B = np.zeros((H, W), dtype=np.uint8)
B[5:10, 5:10] = 255
# 3. B 채널 (Grayscale)
plt.figure(figsize=(20, 4))
plt.imshow(B, cmap='gray')
plt.title('B-Channel')
plt.show()
```

```python
# • 배열을 활용해 이미지 생성
two_channel = np.stack([R, G, np.zeros_like(B)], axis=-1)
# 4. 2채널 (R+G → 노랑 영역)
plt.figure(figsize=(20, 4))
plt.imshow(two_channel)
plt.title('R + G')
plt.show()
```

```python
#  배열을 활용해 이미지 생성
three_channel = np.stack([R, G, B], axis=-1)
# 5. 3채널 (RGB 컬러 전체)
plt.figure(figsize=(20, 4))
plt.imshow(three_channel)
plt.title('R + G + B')
plt.show()
```

```python
image = cv2.imread("C:/Users/PC502/Desktop/download.jpg")
plt.imshow(RGB_image)
plt.show()
```

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 4))
plt.imshow(image)
plt.show()
```

```python
# R, G, B 채널로 분리
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(RGB_image)
```

```python
# 각 채널을 이미지로 표시
plt.figure(figsize=(12, 4)) 
```

```python
# 빨간 채널
plt.subplot(1,3,1)
plt.imshow(R, cmap='Reds')
plt.title('Red Channel')

# 초록 채널
plt.subplot(1,3,2)
plt.imshow(G, cmap='Greens')
plt.title('Green Channel')

# 파란 채널
plt.subplot(1,3,3)
plt.imshow(B, cmap='Blues')
plt.title('Blue Channel')

plt.show()
```

```python
# 이미지 변환 BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# 이미지 변환 RGB to HSV
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
```

```python
# H, S, V 채널로 분리
H, S, V = cv2.split(image_hsv)
```

```python
# 각 채널을 그레이스케일 이미지로 표시
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)

# 색상 
plt.imshow(H, cmap= 'hsv')
plt.title('Hue Channe')

# 채도
plt.subplot(1, 3, 2)
plt.imshow(S, cmap= 'gray')
plt.title('Saturation Channel')

# 명도
plt.subplot(1, 3, 3)
plt.imshow(V, cmap= 'gray')
plt.title('Value Channel') 
plt.show()
```

```python
# 이제 분석을 할 단계 핵심을 파악!
# 전처리를 더 해볼 것이다.
# 이미지의 특징을 "도출" 하는 것이 이미지 전처리의 핵심!
```

```python
# 이미지를 64x64으로 크기 조정
# 10*10으로 하면 알아볼 수 없다 특징을 알아볼 수 있는 선에서 줄이기!
resized_image=cv2.resize(image, (20,20), interpolation=cv2.INTER_AREA)
resized_image=cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
```

```python
# 결과를 시각화하기 위한 코드
plt.figure(figsize=(10,6))
plt.imshow(resized_image)
plt.title('Resize')
plt.show()
```

```python
# 가우시안 블러 적용 5*5
blurred_image = cv2.GaussianBlur(image, (5,5),0)
blurred_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)
```

```python
# 중간값
median_filtered_img = cv2.medianBlur(image, 5)
median_filtered_img = cv2.cvtColor(median_filtered_img, cv2.COLOR_BGR2RGB)
```

```python
# 결과를 시각화하기 위한 코드
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.imshow(blurred_image)
plt.title('Blurred')

plt.subplot(1,2,2)
plt.imshow(median_filtered_img)
plt.title('Median')

plt.show()
```
